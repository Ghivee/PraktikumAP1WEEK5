```python
pertama = float(input('Masukkan Angka Pertama '))     #baris 1
kedua   = float(input('Masukkan Angka Kedua '))       #baris 2
ketiga  = float(input('Masukkan Angka Ketiga '))      #baris 3

if pertama > kedua and pertama > ketiga:              #baris 4
    print(pertama,'Angka Pertama Bilangan Terbesar')  #baris 5
elif kedua > pertama and kedua > ketiga:              #baris 6
    print(kedua,'Angka Kedua Bilangan Terbesar')      #baris 7
else:                                                 #baris 8
    print(ketiga,'Angka Ketiga Bilangan Terbesar')    #baris 9
```

```python
konsumen = input("Masukkan Jenis Konsumen (member/non member): ") #baris 1
belanja = float(input("Masukkan Jumlah Belanja: "))               #baris 2

if (konsumen == "member"):                                        #baris 3
    if (belanja >= 200000):                                       #baris 4
      total = belanja - (belanja * (2/100 + 4/100))               #baris 5
    elif (belanja >= 100000):                                     #baris 6
      total = belanja - (belanja * (2/100 + 3/100))               #baris 7
    else:                                                         #baris 8
      total = belanja - (belanja * 2/100)                         #baris 9
elif (konsumen == "non member"):                                  #baris 10
    if (belanja >= 200000):                                       #baris 11
      total = belanja -(belanja * 4/100)                          #baris 12
    elif (belanja >= 100000):                                     #baris 13
      total = belanja - (belanja * 3/100)                         #baris 14
    else:                                                         #baris 15
      total = belanja                                             #baris 16


print ("Total belanja = "+ str(total))                            #baris 17
```
